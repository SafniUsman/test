```python
# UNQ_C2
# GRADED CELL: Sequential model
tf.random.set_seed(1234) # for consistent results
model = Sequential(
    [               
        Dense(25, input_shape=(400,), activation='relu', name='layer1'),  # layer 1 with 25 units and ReLU activation
        Dense(15, activation='relu', name='layer2'),  # layer 2 with 15 units and ReLU activation
        Dense(10, activation='linear', name='layer3')  # layer 3 with 10 units and linear activation
    ], name="my_model"
)
```

```python
test_model(model, 10, 400)
```

```python
[layer1, layer2, layer3] = model.layers
```

```python
#### Examine Weights shapes
W1,b1 = layer1.get_weights()
W2,b2 = layer2.get_weights()
W3,b3 = layer3.get_weights()
print(f"W1 shape = {W1.shape}, b1 shape = {b1.shape}")
print(f"W2 shape = {W2.shape}, b2 shape = {b2.shape}")
print(f"W3 shape = {W3.shape}, b3 shape = {b3.shape}")
```

```python
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
)

history = model.fit(
    X,y,
    epochs=40
)
```
